5.1 Выберите дом клиенту

5.2 Выберите дом клиенту

Клиент хочет дом либо с очень красивым видом из окна, либо с видом на набережную, в очень хорошем состоянии и год постройки не меньше 1980 года. 
В какой ценовом диапазоне будут дома?

```python
dhouse_for_cust = dhouse[((dhouse['view'] == 4) | (dhouse['waterfront']==1)) & (dhouse['condition']==5) & (dhouse['yr_built'] >= 1980)]
```

```python
dhouse_for_cust
```

```python
dhouse_for_cust['price'].min()
```

```python
dhouse_for_cust['price'].max()
```

**Ценновой дипозон: от 1295000.0 до 3000000.0**

5.3 Выберите дом клиенту
Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. 

Какая оценка по состоянию у таких домов в среднем?
