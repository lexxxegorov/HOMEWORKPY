1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома
Если реновации дома не было, то в новом признаке поставьте 0

```python
dhouse['delta_renovated'] = dhouse.yr_renovated	- dhouse.yr_built
```

```python
dhouse.loc[(dhouse.delta_renovated<0, 'delta_renovated')] = 0
```

```python
dhouse['delta_renovated']
```

1.3 Создайте признаки года продажи, месяца продажи

```python
dhouse['date'] = pd.to_datetime(dhouse['date'])
```

```python
dhouse['year_sale'] = dhouse.loc[:, 'date'].dt.year
```

```python
dhouse['month_sale'] = dhouse.loc[:, 'date'].dt.month
```

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
dhouse.pivot_table(index = 'condition', values = 'price', aggfunc = {'price':['min', 'max', 'mean']})
```
