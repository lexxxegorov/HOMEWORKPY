##Задание 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import seaborn as sns
import numpy as np
corr_matrix = dhouse.corr()
corr_matrix=np.round (corr_matrix, 1)
corr_matrix[np.abs(corr_matrix)<0.3]=0
plt.figure(figsize=(15,10))

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Стоимсоть жилья в больше степени зависит от:
1. Жилой площади
2. Кол-во ванн
3. От категории дома
4. От площади дома за вычетом подвальной
5. от вида

**Зависимость стоимости от размера жилой площади дома**

```python

plt.figure(figsize=(150,150))
sns.jointplot(x=dhouse['price'], y=dhouse['sqft_living'], data=dhouse, kind='reg');
plt.title('Зависимость стоимости от размера жилой площади дома');
```

**Зависимость стоимости от категории дома и площади дома за вычетом подвальной.**

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.scatterplot(
    x='price',
    y='sqft_above',
    data=dhouse,    
    hue='grade',
    size='grade'

)

sns.despine()
```

На диаграмме можем видеть прямую зависимость стоимости жилья от общей площади и категории дома. Определенные категории домов начинаются с определного метража и соотвественно стоимости. 

**Зависимость стоимости от вида.**

```python
plt.figure(figsize=(15,10))
sns.boxplot(x=dhouse['view'], y=dhouse['price'])
plt.title('Зависимость стоимости от вида')
plt.xlabel('Вид')
plt.ylabel('Цена');

```

На данном графике можем видеть зависимость стоимость от вида из окон дома. Чем лучше вид, тем соотвественно и выше стоимость. Также можем видеть, что выборка по стоимости домов с видом 1-4 начинается приблизительно с одной стоимости. Также можно обратить внимание, что дома с видом 1 и 2 в одном ценновом сигменте. Можем сравнить выборку домов по медиане и квантилям, стоимость в 1 000 000 уе, это приблизительно 50% домов с видом 3 и 25% с видом 4, тем самым понимаем, что за данную сумму домов с видом 3 для выбора будет больше.
